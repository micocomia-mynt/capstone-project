### Data Cleaning

```python
# FUNCTIONS

# Function for cleaning appstore df
def clean_app_store_df(df):
  clean_df = df.copy()   #make a copy of the df 
  clean_df = clean_df.drop(['userName', 'app_name', 'app_id', 'developerResponse', 'isEdited'], axis=1) # drop unwanted columns 
  clean_df['Dates'] = pd.to_datetime(clean_df['date']).dt.date # Format date 
  clean_df['Time'] = pd.to_datetime(clean_df['date']).dt.time  # Format time
  clean_df = clean_df.drop('date', axis=1)
  clean_df['final_review'] = clean_df.title.map(str) + " " + clean_df.review # Create final review column (join heading + content)
  clean_df = clean_df.drop(['title', 'review'], axis=1)
  clean_df = clean_df.dropna(subset=['final_review']) 
  return clean_df

# Function for cleaning playstore df
def clean_play_store_df(df):
  clean_df = df.copy()   #make a copy of the df 
  clean_df = clean_df.drop(['_id', 'app_id', 'app_name', 'replyContent', 'reviewCreatedVersion', 'reviewId', 'userImage', 'userName'], axis=1) # drop unwanted columns 
  clean_df['Dates'] = pd.to_datetime(clean_df['at']).dt.date # Format date 
  clean_df['Time'] = pd.to_datetime(clean_df['at']).dt.time  # Format time
  clean_df = clean_df.drop('at', axis=1)
  clean_df.rename(columns={'content':'final_review','score':'rating' }, inplace=True)
  clean_df = clean_df.dropna(subset=['final_review']) 
  return clean_df

# Function to filter df by this timeframe: 2020-1-1 to 2021-5-31
def filter_timeframe_df(df): 
  df = df[(df['Dates']>datetime.date(2020,1,1)) & (df['Dates']<datetime.date(2021,5,31))] #filter
  df = df.reset_index(drop=True) #drop index
  return df 
```

### Ranking mentions of different GCash Services in App Store and PlayStore Reviews

In this section, we'll try to figure out which GCash services are being talked about / mentioned in the App Store and Play Store Reviews. 
Requirements for this section:
1. List of GCash Services -> Check Good News Day slides

#### Preprocessing

Requirements:
1. List of GCash services and possible permutations
2. lowercase, remove punctuation
3. Functionalize str.contains / whatever way you decide to detect

##### FUNCTIONS

```python
# FUNCTIONS
import re

#1. Function to text clean the df
def gcash_services_cleaning(df, column):
  df = df.drop_duplicates()
  df[column] = [entry.lower() for entry in df[column]] # Lowercase text
  df[column] = df[column].apply(lambda x: contractions.fix(x)) # Expand contractions
  df[column] = df[column].replace(r'\r+|\n+|\t+',' ', regex=True) # remove \r, \n, \t
  df[column] = df[column].replace('\s+', ' ', regex=True) # remove multiple white spaces
  df[column] = df[column].str.replace(r'[^\w\d\s\-\+]+', '') # remove punctuation except for dashes
  return df

#2. Function to count word occurences for feature
## FOR GINVEST -> CHEck out seedbox, atram etc etc
# Define terms

terms = {'GInvest':"(ginvest|g invest|g-invest|atram|seed box)",
'GForest':'(gforest|g forest|forest|g-forest|energy|trees|tree)',
'GLife':'(glife|g life|g-life)',
'Send_Money':'(send money|send-money|sending money|clip|clips|send clip|send clips)',
'Bank_Transfer':'(bank transfer|bank transfers|bank-transfer|bank-transfers)',
'Cash_In':'(cash in|cash ins|cashin|cash-in|cash-ins|cashing in)',
'GSave':'(gsave|g save|g-save|cimb)',
'Buy_Load':'(buy load|buying load|buy loads|load rebate|load rebates|load)',
'A+_Rewards':'(a\+|a\+ reward|a\+ rewards)',
'Pay_Bills':'(pay bills|pay bill|paying bills|paying bill|bills)',
'GInsure':'(insurance|insured|ginsure|g-insure|g insure|insured)',
'Cash_Out':'(cash out|cash outs|cashout|cash-out|cash-outs|cashing out)',
'GCash_MasterCard':'(mastercard|master card|gcash mastercard|gcash card|master-card)',
'Donations':'(donate|donating|donation|donations)',
'GCredit':'(credit|gcredit|g credit|g-credit|credit score)',
'Payroll':'(payroll|payrolls|pay roll|pay rolls)',
'QR_Codes':'(qr|qr code|q-r|q r|scan to pay|scan for payments)',
'Remittance':'(remittance|remitting|padala|magpadala|nagpapadala|pagpapadala|western union|moneygram|skrill|cross remittance|alipayhk|sbi remit|emq send|wall street exchange|instant cash|payit|denarii|denarii cash|rocket remit|azimo)',
'Pay_Online':'(online|pay online|paying online)'
}
# count occurences of words defined in the terms dictionary
def count_gcash_feature(df,column,terms):
  for k in terms.keys():
    df[k] = 0
    for words in re.sub("[()]","",terms[k]).split('|'):
        mask = df[column].str.contains(words, case = False)
        df[k] += mask
  # Clip Feature count columns == Ensure that rows are just tagged as 1
  df[feature_list] = df[feature_list].clip(0,1)
  #split df into three parts
  df_good = df.loc[df['rating'].isin([4,5])].copy()
  df_bad = df.loc[df['rating'].isin([1,2,3])].copy()
  return df, df_good, df_bad

def time_filter_count_gcash_feature(df,column,terms,year,month):
  for k in terms.keys():
    df[k] = 0
    for words in re.sub("[()]","",terms[k]).split('|'):
        mask = df[column].str.contains(words, case = False)
        df[k] += mask
  # Clip Feature count columns == Ensure that rows are just tagged as 1
  df = df.loc[(pd.to_datetime(df['Dates']).dt.month == month) & (pd.to_datetime(df['Dates']).dt.year == year) ]
  df[feature_list] = df[feature_list].clip(0,1)
  #split df into three parts
  df_good = df.loc[df['rating'].isin([4,5])].copy()
  df_bad = df.loc[df['rating'].isin([1,2,3])].copy()
  return df, df_good, df_bad


# 3. Function that sums and outputs a dataframe
def sum_column_features(df):
  # Build a dictionary from sum of pandas column
  feature_list = ['GInvest', 'GForest', 'GLife', 'Send_Money', 'Bank_Transfer', 'Cash_In', 'GSave', 'Buy_Load', 'A+_Rewards', 'Pay_Bills', 'GInsure', 'Cash_Out', 'GCash_MasterCard', 'Donations', 'GCredit', 'Payroll', 'QR_Codes', 'remittance', 'Pay_Online']
  df = df.filter(items = feature_list).copy() #filter columns based on feature list
  return df.sum(axis=0).to_dict() #return a dictionary of sum of outputs

# 4. Function to export dictionary to CSV (For Power BI plotting)

def export_to_csv(dict, filename):
  with open(filename, 'w', newline='') as csvfile:
    fieldnames= ['GCash Features', 'count']
    writer = csv.DictWriter(csvfile,fieldnames=fieldnames)

    writer.writeheader()
    for key in dict:
      writer.writerow({'GCash Features': key, 'count': dict[key]})


def export_to_csv_many(list_dict, filename,header_list):
  with open(filename,'w', newline='') as f:
    writer = csv.writer(f, delimiter='\t')
    writer.writerow(header_list)
    for key in list_dict[0]:
      writer.writerow([key] + [d[key] for d in list_dict])

```
