```python
df_appstore_copy = df_appstore.copy()
```

```python
df_appstore_copy
```

```python
df_appstore_copy['combined_text'] = df_appstore_copy['title'] + ' ' + df_appstore_copy['review']
df_appstore_copy
```

```python
prediction1 = []
prediction2 = []
confidence_level1 = []
confidence_level2 = []

for line in df_appstore_copy['combined_text']:
    line = line.replace("\n"," ")
    prediction = model.predict(line.lower(), k=2)
    prediction_lang1 = prediction[0][0]
    prediction_lang2 = prediction[0][1]
    confidence1 = prediction[1][0]
    confidence2 = prediction[1][1]
    prediction1.append(prediction_lang1)
    confidence_level1.append(confidence1)
    prediction2.append(prediction_lang2)
    confidence_level2.append(confidence2)

# you add the list to the dataframe, then save the datframe to new csv
df_appstore_copy['prediction1'] = prediction1
df_appstore_copy['confidence_level1'] = confidence_level1
df_appstore_copy['prediction2'] = prediction2
df_appstore_copy['confidence_level2'] = confidence_level2
```

```python
df_appstore_copy.head()
```

```python
df_appstore_copy['prediction1'].value_counts()
```

```python
df_appstore_copy['prediction2'].value_counts()
```

```python
mask = (df_appstore_copy['prediction1'] == '__label__tl') | \
        ((df_appstore_copy['prediction2'] == '__label__tl') & (df_appstore_copy['prediction1'] == '__label__en') \
        & (df_appstore_copy['confidence_level1'] <= 0.9))
df_appstore_tl = df_appstore_copy[mask]
```

```python
df_appstore_tl.head(20)
```

```python
df_appstore_en = df_appstore_copy[(df_appstore_copy['prediction1'] == '__label__en') & (df_appstore_copy['confidence_level1'] >= 0.9)]
```

```python
df_appstore_en.head(20)
```

```python
df_appstore_tl.to_csv('data/df_appstore_tl.csv')
df_appstore_en.to_csv('data/df_appstore_en.csv')
```
